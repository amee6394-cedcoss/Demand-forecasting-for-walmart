```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('walmart.csv')  #for reading data
df  
```

```python
y=df['weeklySales']
x=df.drop(['weeklySales','Type'],axis=1)  #droping weeklysales temporary
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3) #spliting data 
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True) #droping column parmanantly
```

```python
df.columns #showing column's names
```

```python
m=RandomForestRegressor() 
```

```python
m.fit(x_train,y_train) #this function is used for ploting line between x_train and y_train
```

```python
p=m.predict(x_test) # predict is used for predicting value which is store in p
```

```python
sc=m.score(x_test,y_test)
sc*100
```
